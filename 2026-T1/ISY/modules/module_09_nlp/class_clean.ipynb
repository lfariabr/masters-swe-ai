```python
import re

random_text_to_be_cleaned = "This is a random text that @!@!#!# SHOULD be cleaned <!# 18293812931>😊"

def clean_text(text):
    # remove special characters first
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    print(f"After removing special characters:\n{cleaned_text}\n")
    # remove emojis
    cleaned_text = re.sub(r'[^\x00-\x7F]+', '', cleaned_text)
    print(f"After removing emojis:\n{cleaned_text}\n")
    # remove numbers
    cleaned_text = re.sub(r'\d+', '', cleaned_text)
    print(f"After removing numbers:\n{cleaned_text}\n")
    return cleaned_text


cleaned_text = clean_text(random_text_to_be_cleaned)
print(cleaned_text)
```

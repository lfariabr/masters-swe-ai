```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Job Position': ['CEO', 'CTO', 'Junior Manager', 'Employee'],
    'Years of Experience': [20, 15, 5, None],
    'Salary': [200000, 150000, None, 30000]
})

df
```

```python
# Interpolating missing values in the 'Years of Experience' column
df['Years of Experience'] = df['Years of Experience'].interpolate()
df['Salary'] = df['Salary'].interpolate()

df
```

## Encoding: One-Hot Encoding on data


```python
df1 = pd.get_dummies(df['Job Position'])
df1
```

```python
df = pd.concat([df, df1], axis=1).reindex(df.index)
df
```

```python
# Removing the column Job position from df as it is of no use now
df.drop('Job Position', axis=1, inplace=True)
df
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

# creating instance
le = LabelEncoder()

# using fit_transform to encode the 'Job Position' column
label = le.fit_transform(df['Job Position'])

# 
label
```

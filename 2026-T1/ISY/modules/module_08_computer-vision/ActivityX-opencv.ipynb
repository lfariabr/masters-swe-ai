# Open CV walkthrough

```python
import cv2
from cv2 import CascadeClassifier, imread
```

```python
# cv2.data.haarcascades points to OpenCV's bundled XML files — no manual download needed
xml_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
classifier = CascadeClassifier(xml_path)

# test1.jpg lives in the same folder as this notebook
pixels = imread('test1.jpg')
print(pixels)

# perform face detection
bboxes = classifier.detectMultiScale(pixels)

# print bounding box for each detected face
for box in bboxes:
    print(box)
```

# MTCNN wallkthrough

```python
from facenet_pytorch import MTCNN
from PIL import Image

# facenet-pytorch MTCNN expects a PIL image
img = Image.open('test1.jpg')

detector = MTCNN()
boxes, probs = detector.detect(img)

print('Bounding boxes:', boxes)
print('Confidences:  ', probs)
```

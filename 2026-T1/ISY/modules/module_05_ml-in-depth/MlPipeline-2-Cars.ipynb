```python
# ===================== SUMMARY =====================
# This code:
# 1. Imports required libraries (NumPy, Pandas, Math)
# 2. Configures Pandas display settings for better readability
# 3. Uploads a CSV file in Google Colab
# 4. Reads the car dataset into a DataFrame
# 5. Defines column (feature) names for the dataset
# 6. Reloads the dataset with proper column headers
# 7. Displays the first 2 rows of the dataset
# ===================================================


# Import necessary libraries for data handling and numerical operations
import numpy as np
import pandas as pd
import math


# Configure pandas display settings:
# - Show floats with 1 decimal place
# - Limit output to 15 rows for readability
pd.options.display.float_format = '{:.1f}'.format
pd.options.display.max_rows = 15


# Upload file manually in Google Colab environment
car_data_csv = '/Users/luisfaria/Desktop/sEngineer/masters_SWEAI/2026-T1/ISY/modules/module_05_ml-in-depth/cars_data.csv'

# Initial read of dataset (without column names)
car_data = pd.read_csv(car_data_csv)


# Define column names based on dataset documentation
feature_names = ['symboling', 'normalized-losses', 'make', 'fuel-type',
        'aspiration', 'num-doors', 'body-style', 'drive-wheels',
        'engine-location', 'wheel-base', 'length', 'width', 'height', 'weight',
        'engine-type', 'num-cylinders', 'engine-size', 'fuel-system', 'bore',
        'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
        'highway-mpg', 'price']


# Reload dataset with proper column names and encoding
car_data = pd.read_csv(
    "cars_data.csv",
    names=feature_names,
    header=None,
    encoding='latin-1'
)


# Display first 2 rows to verify data structure
car_data.head(2)
```

```python
# ===================== SUMMARY =====================
# This code:
# 1. Randomly shuffles (reorders) the dataset rows
# 2. Helps prevent any ordering bias that could affect model training
# 3. Prints the total number of records in the dataset
# 4. Displays the first 10 rows after shuffling
# ===================================================


# Shuffle the dataset randomly to avoid any ordering bias
#car_data = car_data.reindex(np.random.permutation(car_data.index))
car_data = car_data.sample(frac=1, random_state=42)


# Print total number of instances (rows) in the dataset
print(" Instances of car prices and details loaded:  ", len(car_data))


# Display first 10 rows of the shuffled dataset
car_data.head(10)
```

```python
# ===================== SUMMARY =====================
# This code:
# 1. Defines the target variable (label) as 'price'
# 2. Separates numeric features from the dataset
# 3. Identifies categorical features by excluding numeric features and the label
# 4. Verifies the number of numeric and categorical features using assertions
# ===================================================


# Define the target variable (what we want to predict)
LABEL = 'price'


# Select numeric feature columns from the dataset
numeric_feature_names = car_data[['symboling', 'normalized-losses', 'wheel-base',
        'length', 'width', 'height', 'weight', 'engine-size', 'horsepower',
        'peak-rpm', 'city-mpg', 'highway-mpg', 'bore', 'stroke',
         'compression-ratio']].columns.tolist()


# Identify categorical features by removing numeric features and label from all features
categorical_feature_names = list(set(feature_names) - set(numeric_feature_names) - set([LABEL]))


# Print number of numeric features
print(len(numeric_feature_names))


# Validate expected counts of numeric and categorical features
assert len(numeric_feature_names) == 15
assert len(categorical_feature_names) == 10
```

```python
# Convert numeric features and label to numeric type (invalid values become NaN)
for feature_name in numeric_feature_names + [LABEL]:
    car_data[feature_name] = pd.to_numeric(car_data[feature_name], errors='coerce')

# Count missing values in each column
missing_values = car_data.isna().sum()

# Get only columns that have at least one missing value
columns_with_missing = missing_values[missing_values > 0].index.tolist()

# Print the names of columns with missing values
print("Columns with missing values:", columns_with_missing)

# Optionally fill missing values with 0 (commented out for now)
# car_data.fillna(0, inplace=True)

# Store original dataset length before cleaning
original_length = len(car_data)

# Drop 'normalized-losses' column due to excessive missing values
car_data.drop(labels='normalized-losses', axis=1, inplace=True)

# Update numeric features list
numeric_feature_names = list(set(numeric_feature_names) - set(['normalized-losses']))

# Drop all rows that still contain missing values
car_data.dropna(inplace=True)

# Store new dataset length after cleaning
new_length = len(car_data)

# Print summary of rows dropped and final dataset size
print('Dropped {} rows with missing data'.format(original_length - new_length))
print('Remaining rows:', len(car_data))
```

```python
# ===================== SUMMARY =====================
# This code:
# 1. Imports necessary ML libraries for regression, preprocessing, and model evaluation
# 2. Defines input features (numeric columns) and target variable ('price')
# 3. Optionally includes polynomial feature expansion (commented out)
# 4. Splits the dataset into training (80%) and testing (20%) sets
# 5. Initializes four regression models:
#    - LinearRegression (LR)
#    - GradientBoostingRegressor (GBR)
#    - HistGradientBoostingRegressor (HGBR)
#    - RandomForestRegressor (RBR)
# 6. Loops through each model:
#    - Fits the model on training data
#    - Predicts target values on test data
#    - Computes RÂ² score to evaluate model performance
#    - Prints the model name and its RÂ² score
# ===================================================

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import (GradientBoostingRegressor, HistGradientBoostingRegressor, RandomForestRegressor)

# Select numeric features as inputs (X) and 'price' as target (y)
x = car_data[numeric_feature_names]
y = car_data['price']

# Optional polynomial feature expansion (currently commented out)
# poly = PolynomialFeatures()
# x = poly.fit_transform(x)

# Split data into training (80%) and testing (20%) sets
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2
)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize regression models
LR = LinearRegression()
GBR = GradientBoostingRegressor()
HGBR = HistGradientBoostingRegressor()
RBR = RandomForestRegressor()

n_jobs = -1  # Not used here, but could be passed to parallelizable models

# Fit each model, predict on test set, and evaluate RÂ² score
for model in [LR, GBR, HGBR, RBR]:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    r2 = r2_score(y_test, y_pred)
    print(model, r2)
```

```python
# ===================== SUMMARY =====================
# This code:
# 1. Imports RandomForestRegressor and RÂ² metric for evaluation
# 2. Performs manual hyperparameter tuning for Random Forest by looping over:
#    - max_depth: [10, 20, 30]
#    - n_estimators (number of trees): [200, 250, 300, 350, 400]
#    - min_samples_leaf (minimum samples per leaf): [1, 2, 4]
# 3. For each combination:
#    - Initializes a RandomForestRegressor with the current hyperparameters
#    - Trains the model on the training set
#    - Predicts target values on the test set
#    - Computes RÂ² score to measure model performance
#    - Prints the hyperparameter combination and corresponding RÂ² score
# ===================================================

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Hyperparameters
for depth in [10, 20, 30]:  # max_depth
    for n in [200, 250, 300, 350, 400]:  # n_estimators
        for leaf in [1, 2, 4]:  # min_samples_leaf

            # Initialize Random Forest with current hyperparameters
            model = RandomForestRegressor(
                n_estimators=n,
                max_depth=depth,
                min_samples_leaf=leaf,
                n_jobs=-1,
                random_state=42
            )

            # Train model on training data
            model.fit(x_train, y_train)

            # Predict on test data
            y_pred = model.predict(x_test)

            # Compute RÂ² score
            r2 = r2_score(y_test, y_pred)

            # Print hyperparameters and model performance
            print(f"n_estimators={n}, max_depth={depth}, min_leaf={leaf}, R2={r2:.4f}")
```

```python
# ===================== SUMMARY =====================
# This code:
# 1. Performs hyperparameter tuning for RandomForestRegressor by iterating over:
#    - max_depth: [10, 20, 30]
#    - n_estimators: [200, 250, 300, 350, 400]
#    - min_samples_leaf: [1, 2, 4]
# 2. For each combination:
#    - Initializes and trains the Random Forest model
#    - Predicts on the test set
#    - Computes RÂ² score to evaluate model performance
# 3. Keeps track of the best-performing model by:
#    - Comparing current RÂ² score with the best_score so far
#    - Updating best_score and best_params when a better score is found
# 4. Prints the RÂ² score for each combination
# 5. After all iterations, prints the best RÂ² score and corresponding hyperparameters
# ===================================================

best_score = -1
best_params = {}

for depth in [10, 20, 30]:
    for n in [200, 250, 300, 350, 400]:
        for leaf in [1, 2, 4]:

            # Initialize Random Forest with current hyperparameters
            model = RandomForestRegressor(
                n_estimators=n,
                max_depth=depth,
                min_samples_leaf=leaf,
                n_jobs=-1,
                random_state=42
            )

            # Train the model
            model.fit(x_train, y_train)

            # Predict on test data
            y_pred = model.predict(x_test)

            # Compute RÂ² score
            r2 = r2_score(y_test, y_pred)

            # Update best score and parameters if current model is better
            if r2 > best_score:
                best_score = r2
                best_params = {
                    'n_estimators': n,
                    'max_depth': depth,
                    'min_samples_leaf': leaf
                }

            # Print current hyperparameters and RÂ²
            print(f"n={n}, depth={depth}, leaf={leaf}, R2={r2:.4f}")

# Print the best RÂ² score and corresponding hyperparameters
print("\nBest Score:", best_score)
print("Best Params:", best_params)
```

```python
# ===================== SUMMARY =====================
# This code:
# 1. Imports MinMaxScaler from sklearn for feature normalization
# 2. Initializes the scaler object
# 3. Applies Min-Max scaling to input features (x) so that all values are in the range [0, 1]
# 4. Converts the scaled array back to a Pandas DataFrame with original column names
# 5. Prepares the target variable (y_series) from the 'price' column
# 6. Resets the indices of both x and y_series to ensure alignment after scaling
# 7. Displays the first 10 rows of the normalized input features
# ===================================================

from sklearn.preprocessing import MinMaxScaler

# Instantiate the scaler
scaler = MinMaxScaler()

# Apply the scaler to numeric features
X_normalized = scaler.fit_transform(x)

# Convert scaled array back to DataFrame
x = pd.DataFrame(X_normalized, columns=x.columns)

# Prepare target variable
y_series = car_data['price']

# Reset indices to align X and y
x.reset_index(drop=True, inplace=True)
y_series.reset_index(drop=True, inplace=True)

# Show first 10 rows of normalized features
x.head(10)
```

```python
# ===================== SUMMARY =====================
# This code:
# 1. Splits the dataset into training (80%) and testing (20%) sets using train_test_split
# 2. Performs manual hyperparameter tuning for RandomForestRegressor by iterating over:
#    - max_depth: [10, 20, 30]
#    - n_estimators: [200, 250, 300, 350, 400]
#    - min_samples_leaf: [1, 2, 4]
# 3. For each combination of hyperparameters:
#    - Initializes and trains the Random Forest model on the training set
#    - Predicts target values on the test set
#    - Calculates RÂ² score to evaluate model performance
#    - Updates best_score and best_params if current RÂ² is higher than previous best
#    - Prints current hyperparameters and corresponding RÂ² score
# 4. After all iterations, prints the best RÂ² score and the hyperparameters that achieved it
# ===================================================

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Split features and target into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2
)

best_score = -1
best_params = {}

# Hyperparameter grid search
for depth in [10, 20, 30]:
    for n in [200, 250, 300, 350, 400]:
        for leaf in [1, 2, 4]:

            # Initialize Random Forest with current hyperparameters
            model = RandomForestRegressor(
                n_estimators=n,
                max_depth=depth,
                min_samples_leaf=leaf,
                n_jobs=-1,
                random_state=42
            )

            # Train the model
            model.fit(x_train, y_train)

            # Predict on test data
            y_pred = model.predict(x_test)

            # Compute RÂ² score
            r2 = r2_score(y_test, y_pred)

            # Update best score and parameters if current model is better
            if r2 > best_score:
                best_score = r2
                best_params = {
                    'n_estimators': n,
                    'max_depth': depth,
                    'min_samples_leaf': leaf
                }

            # Print current hyperparameters and RÂ² score
            print(f"n={n}, depth={depth}, leaf={leaf}, R2={r2:.4f}")

# Print the best RÂ² score and corresponding hyperparameters
print("\nBest Score:", best_score)
print("Best Params:", best_params)
```
